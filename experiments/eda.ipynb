```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATA_FOLDER = '../data/advertising.csv'

data = pd.read_csv(DATA_FOLDER)
```

```python
sns.histplot(data['Age'], bins=20)
```

```python
sns.histplot(data['Daily Internet Usage'], bins=20)
```

```python
grouped_data = data.groupby('Male')['Clicked on Ad'].sum()

labels = ['Female (0)', 'Male (1)']
values = grouped_data.values

plt.pie(
    values,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
)
plt.title('Proportion of Clicked Ads by Gender')
plt.show()
```

```python
corr = data.corr(numeric_only=True)
plt.figure(figsize=(12,7))
data_plot = sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
